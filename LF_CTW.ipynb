**<h2>과제 1</h2>**

스크래핑 대상 사이트

https://everycoding.net/static/upload/tutor/image/value_investment/sample_site.html

위의 sample 사이트에서 everycoding의 적정주가를 산출하는 코드를 작성합니다.

PER x EPS = 적정주가

<h3>프로젝트 내용 설명</h3>

**[과제 프로젝트 강좌](https://everycoding.net/tutor/43)**



```
# 1. 필요한 모듈을 import 합니다.
import requests
import pandas

# 2. 함수를 정의합니다.
def getEverycodingFairPrice():

  url="https://everycoding.net/static/upload/tutor/image/value_investment/sample_site.html"
  
  # @@@@아래 내용을 수정하세요@@@@
  # 3. url 값을 이용하여 사이트의 html을 문자열 형태로 가져옵니다.
  html = requests.get(url)
  html_text = html.text
  
  
  # @@@@아래 내용을 수정하세요@@@@
  # 4. html 형식의 문자열들을 pandas를 이용해 리스트 Dataframe(table) 형태로 변환합니다. (parse)
  #힌트 : pandas.read_html(...)를 사용하세요!
  
  tables = pandas.read_html(html_text)
  
  
  # @@@@아래 내용을 수정하세요@@@@
	# 5. 사이트의 재무제표들중 Every 주식의 재무제표의 테이블들의 값을 가져옵니다.
  
  # 힌트 : 몇 번째 표가 Every 주식일까요?
	# 힌트 : 값을 가져오기 위해서는 어떤 코드를 작성하였나요? values
  
  everycoding_stock = tables[1].values
  
	# 6. 재무제표 테이블에서 PER * EPS = 적정주가 공식을 이용하여 적정주가를 계산합니다.
  # per = everycoding_stock[?][?]
  # eps = everycoding_stock[?][?]
  # ? 안에 들어가야 하는 값은 무엇일까요?
  
  per = float(everycoding_stock[2][1])
  eps = float(everycoding_stock[3][1])  
  
  # @@@@아래 내용을 수정하세요@@@@
  # PER x EPS = 적정주가 공식을 적용해보세요!
  fair_price = per * eps
  
  # 7. 계산하여 나온 값을 return 합니다.
  return fair_price

# 정답을 확인해 볼까요?
print(getEverycodingFairPrice())



# 2019.05.21일 기준 대략적인 값은 24837.76 입니다.
# 값이 계속 최신화 되어 정답이 바뀝니다.

```

**<h2>과제 4</h2>**

스크래핑 대상 사이트

http://media.kisline.com/highlight/mainHighlight.nice?paper_stock=005930&nav=1

현재 주식의 가격을 입력하면 

상장온라인 사이트에서 현재 주식의 가격과, 산출한 적정 주가를 비교하는 함수를 작성합니다.

PER x EPS = 적정주가

<h3>프로젝트 내용 설명</h3>

**[과제 프로젝트 강좌](https://everycoding.net/tutor/43/lesson/172/view#/1)**






```
# 1. 필요한 라이브러리들을 import 합니다.
import requests
import pandas


# 과제 3의 함수입니다.

# 2. 크롤링 대상 사이트의 html을 문자열 형태로 가져옵니다.
def getFairPriceByCompanyCode(companyCode):

    url = 'http://media.kisline.com/highlight/mainHighlight.nice?paper_stock='+str(companyCode)+'&nav=1'
    
    # @@@@아래 내용을 수정하세요@@@@
    html = requests.get(url)
    

    # @@@@아래 내용을 수정하세요@@@@
    # 3. html 형식의 문자열들을 pandas를 이용해 리스트 Dataframe(table) 형태로 변환합니다. (parse)
    tables = pandas.read_html(html.text)
    
    
    # 4. 보여지는 재무제표 값들을 가져옵니다.
    financialSummary=tables[6].values
    
    # 5. PER 값을 가져옵니다.
    per = float(financialSummary[10][4])
    # 6. EPS 값을 가져옵니다.
    eps = float(financialSummary[13][4])
    
    
    
    
    
    # @@@@아래 내용을 수정하세요@@@@
    # 7. PER x EPS = 적정주가 라는 공식을 대입합니다.
    fairPrice = per*eps
    return fairPrice

   
  

  
def feedbackStock(companyCode,nowPrice):
    # getFairPriceByCompanyCode(...) 는 기업의 코드를 입력받으면 적정주가를 계산하는 함수
    fairPrice = getFairPriceByCompanyCode(companyCode)
    
    
    # @@@@아래 내용을 수정하세요@@@@
    # 만약 nowPrice > fairPrice 이면 '고 평가'
    if nowPrice > fairPrice:
        print('고평가')

      
     # @@@@아래 내용을 수정하세요@@@@
     # 그렇지 않고 nowPrice < fairPrice 이면 '저 평가'
    elif nowPrice < fairPrice:
        print('저평가')

    
    # 전부다 아니면 동일한 가격
    else:
        print('동일한 가격')
        
    print('fairPrice : ',fairPrice,' nowPrice : ',nowPrice)

  
  
feedbackStock('035720',124500)



# 정답은 '고평가' 입니다. 
# 이 값이 나오도록 코드를 작동시키세요.
```
